## Menggunakan format percent untuk tipedata float
langkah ini bersifat opsional, hanya untuk memudahkan membaca dataframe<\br>

ref: https://stackoverflow.com/questions/23981601/format-certain-floating-dataframe-columns-into-percentage-in-pandas</br>
doc: https://pandas.pydata.org/pandas-docs/version/0.18/generated/pandas.formats.style.Styler.format.html

````python
trx_year_month = df.groupby(by = ['year','month'],
    as_index = False).agg(
    Unique_Trx_YM = ('Order_ID', 'nunique'),
    Avg_Discount_YM = ('Discount', 'mean'),
    Avg_Profit_Per_YM = ('Profit', 'mean'))
# need to format to percentage for easier read
trx_year_month['Avg_Trx'] = trx_year_month['Unique_Trx_YM'].mean()

trx_year_month['Avg_Discount'] = df['Discount'].mean()

trx_year_month['Avg_Discount%_YM'] = pd.Series(["{0:.2f}%".format(val * 100) for val in trx_year_month['Avg_Discount_YM']],
                                             index = trx_year_month.index)

trx_year_month['Avg_Discount%'] = pd.Series(["{0:.2f}%".format(val * 100) for val in trx_year_month['Avg_Discount']],
                                             index = trx_year_month.index)

trx_year_month['Avg_Profit'] = df['Profit'].mean()

'''
Menyusun column sesuai kebutuhan, dengan cara:
cols = list(trx_year_month.columns.values)
display(cols)
kemudian paste output code tsb
'''
trx_year_month = trx_year_month[['year',
                                'month',
                                'Unique_Trx_YM',
                                'Avg_Trx',
                                'Avg_Discount_YM',
                                'Avg_Discount',
                                'Avg_Discount%_YM',
                                'Avg_Discount%',
                                'Avg_Profit_Per_YM',
                                'Avg_Profit']]

display(trx_year_month)

# 1. bool 'Avg_Discount%_YM' > 'Avg_Discount%', true or false v
# 2. Profit agg
# 3. Profit mean
# 4. bool Profit agg > Profit mean, true or false
# 5. bool true and true = true, else false
# 6. count or pie diagram last bool
# dibawah aja biar rapih? atau satuin disini
````
